# If

```python
age = 22

print(age >= 18)

if age >= 18:
    print(age, ' is above 18')
```

# If & Else

```python
age = 12

print(age >= 18)

if age >= 18:
    print(age, ' is above 18')
else:
    print(age, ' is below 18')
    
```

# Using Logical Operators

```python
username = input('Enter your User Name : ')

password = input('Enter your Password : ')

print('C1 : ', (username == 'admin'))

print('C2 : ', (password == 'admin123'))

print('----------------------------')

if ((username == 'admin') and (password == 'admin123')):
    print('Welcome User')
else:
    print('Invalid Credentials!...')
```

# If & Elif & Else

```python
n = int(input('Enter a Value : '))

if n > 0 :
    print(n, ' is +ve')
elif n < 0:
    print(n, ' is -ve')
else:
    print(n, 'is Zero')
```

# Nested  If

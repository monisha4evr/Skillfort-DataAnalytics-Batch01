## Disadvantage:   
- Slow Execution Speed:
- Dynamic Typing
- Everything is an Object
- Garbage Collection

    **Partially interpreter**
    Python is called partially interpreted because Python source code is first compiled into bytecode, and then the Python Virtual Machine interprets and executes that bytecode line by line.

- Python is slower because it is an interpreted and dynamically typed language.
It performs runtime type checking, object handling, and garbage collection, which add extra processing overhead.

- High Memory Consumption:
    Python consumes more memory because everything in Python is treated as an object.
Each object stores additional information like datatype, reference count, and metadata.
Python also uses dynamic typing and garbage collection, which require extra memory management.

# Immutable Core Types
    Once created, the value of these objects cannot be changed in memory. 
    If you modify them, Python creates a completely new object.
- Integers (int)
- Floats (float)
- Booleans (bool)
- Strings (str) 
- Tuples (tuple) ( **Collection** )
      

# Mutable Core Types
    The value of these objects can be changed directly in memory
    without creating a new object.
**Collection**
1. List (Mutable) 
3. Set- (Mutable) 
4. Dictionary-(Mutable) 

```python
a=[1,2,3,4]
b=[1,2,3,4]
print(id(a),id(b))
```

```python
a=10
print(a)
print(id(a))
a=20
print(a)
print(id(a))
```

```python
a=[1,2,3]
print(a)
print(id(a))
a.append(4)
print(a)
print(id(a))
```
